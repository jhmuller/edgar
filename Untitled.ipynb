```python
import datetime
def get_quarter(dt):
    qtr = int((dt.month-1) / 3) + 1
    return qtr
dts = [datetime.datetime(2021, 1, 1), datetime.datetime(2021, 2, 1),
       datetime.datetime(2021, 3, 1), datetime.datetime(2021, 4, 1),
       datetime.datetime(2021, 5, 1), datetime.datetime(2021, 6, 1),       
       datetime.datetime(2021, 7, 1), datetime.datetime(2021, 8, 1),
       datetime.datetime(2021, 9, 1), datetime.datetime(2021, 10, 1),       
       datetime.datetime(2021, 11, 1), datetime.datetime(2021, 12, 1)]
for dt in dts:
  print(dt, get_quarter(dt))
```

```python
first_day = datetime.date(2021, 4, 1)
last_day = datetime.date(2021,4,20)

years = [2021]
for year in years:
  for qtr in [1]: #2, 3, 4]:
    months = [m + (qtr-1)*3 for m in [1, 2, 3]]
    url_pre = "https://www.sec.gov/Archives/edgar/daily-index/{0}/QTR{1}/".format(year, qtr)
    for month in months:
      last_day = calendar.monthrange(year, month)[1]      
      for day in range(1, 4): # last_day+1):
        url_post = "company.{0}{1}{2}.idx".format(str(year), str(month).zfill(2), str(day).zfill(2))
        url = url_pre  +  url_post
        print(url)
        try:
          with urllib.request.urlopen(url) as f:
            lines = f.read().decode('utf-8').split("\n")
          print("Got {0} lines".format(len(lines)))
          for li in range(20):
            pass
            #print(lines[li])
          flines = [l for l in lines if re.search("13F", l)]
          for fl in flines:
            print(fl)
        except Exception as e:
          extype, exval, tb = sys.exc_info()
          print("{0}, {1}".format(extype, exval))
          
```

```python

savedir = "./data"
if not os.path.isdir(savedir):
  os.makedirs(savedir)
if not os.path.isdir(savedir):
  raise RuntimeError("no dir {0}".format(savedir))
  
# Initialize a downloader instance. If no argument is passed
# to the constructor, the package will download filings to
# the current working directory.
dl = Downloader(savedir)
```
