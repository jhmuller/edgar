```python
dfs = []
for hdir in hdirs:
  files = [x for x in os.listdir(hdir) if x.endswith(".csv")]
  #print(hdir, len(files))
  for fname in files:
    fparts = fname.split("_", maxsplit=3)
    CIK = fparts[1][3:]
    fpath = os.path.join(hdir, fname)
    df = pd.DataFrame()
    try:
      df = pd.read_csv(fpath)        
    except:
      pass
      #print(sys.exc_info())
    if df.shape[0] == 0:
      continue
    df["CIK"] = CIK
    df["manager"] = fparts[0]     
  if df.shape[0] > 0:
    dfs.append(df)
fulldf = pd.concat(dfs)
display(fulldf.shape)
```

```python
fulldf.reset_index(inplace=True, drop=True)
fulldf["value"] = fulldf["value"].astype("float")
fulldf["otherManager"] = fulldf["otherManager"].astype("str")
fulldf["cusip"] = fulldf["cusip"].astype("str")
fulldf["perSh"].fillna(np.nan)
fulldf["filingDt"] = fulldf["filingDt"].apply(lambda x: datetime.datetime.strptime(str(x), "%Y-%m-%d %H:%M:%S"))
#fulldf["sshPrnamt"] = fulldf["sshPrnamt"].astype(int)
dtypes = fulldf.dtypes
for i, (col, value) in enumerate(dtypes.iteritems()):
  #print(i, col, value)
  if value == "int64":
    fulldf[col] = fulldf[col].astype(int)
  if value == "float64":
    fulldf.loc[fulldf[col].apply(lambda x: np.isinf(x)), col] = np.nan
fulldf.dtypes
```
